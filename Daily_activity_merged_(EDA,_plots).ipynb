```python
data = pd.read_csv("dailyActivity_merged.csv")

# Find the unique IDs
unique_ids = data['Id'].unique()

# List of 33 unique Indian names
names = [
    "Aarav", "Aarushi", "Advik", "Ananya", "Arjun", "Avani", "Devika", "Esha", "Ishaan", "Jiya",
    "Kabir", "Kavya", "Lakshya", "Mahi", "Meera", "Nandini", "Nirav", "Prisha", "Raj", "Ria",
    "Rohan", "Riya", "Saanvi", "Shaan", "Shreya", "Siddharth", "Simran", "Tanisha", "Tarun",
    "Trisha", "Vansh", "Vedika", "Yash"
]

# Create a mapping of unique IDs to Indian names
id_to_name_mapping = dict(zip(unique_ids, names))

# Assign Indian names to each ID in the DataFrame
data['Name'] = data['Id'].map(id_to_name_mapping)
```

```python
assigned_names = data.groupby('Id')['Name'].unique()

# Check if any IDs have multiple names assigned
for id, names in assigned_names.items():
    if len(names) != 1:
        print(f"ID {id} has multiple names assigned: {names}")
    else:
        print(f"ID {id}  assigned  name is: {names[0]}")
```

```python
# Save the modified DataFrame back to the original CSV file
data.to_csv("dailyActivity_merged.csv", index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.histplot(data['TotalSteps'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Total Steps')
plt.xlabel('Total Steps')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='IndianName', y='TotalSteps', data=data)
plt.title('Boxplot of Total Steps by Indian Name')
plt.xlabel('Name')
plt.ylabel('Total Steps')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Name', y='SedentaryMinutes', data=data)
plt.title('Comparison of Sedentary Minutes Across  Names')
plt.xlabel('Name')
plt.ylabel('Sedentary Minutes')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
corr_matrix = data[['TotalSteps', 'VeryActiveMinutes', 'Calories', 'SedentaryMinutes']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
calories_burned = data.groupby('Name')['Calories'].sum().reset_index()

most_calories = calories_burned[calories_burned['Calories'] == calories_burned['Calories'].max()]
print("Person with the most calories burned:")
print(most_calories)


least_calories = calories_burned[calories_burned['Calories'] == calories_burned['Calories'].min()]
print("\nPerson with the least calories burned:")
print(least_calories)
```
