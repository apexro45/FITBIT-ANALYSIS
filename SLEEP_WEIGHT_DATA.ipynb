```python
import seaborn as sns
import matplotlib.pyplot as plt

# Summary statistics
print(merged_df.describe())
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(merged_df['TotalMinutesAsleep'], kde=True, color='skyblue')
plt.title('Distribution of Total Minutes Asleep')
plt.xlabel('Total Minutes Asleep')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(15, 6))
sns.boxplot(x='Name', y='TotalTimeInBed', data=merged_df)
plt.title('Box plot of Total Time In Bed by Name')
plt.xlabel('Name')
plt.ylabel('Total Time In Bed')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
correlation_matrix = merged_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
