```python
import matplotlib.pyplot as plt
import seaborn as sns

print(hourly_merged.describe())

# Univariate Analysis
# Histogram of Calories
plt.figure(figsize=(8, 6))
sns.histplot(hourly_merged['Calories'], kde=True, color='skyblue')
plt.title('Distribution of Calories')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
correlation_matrix = hourly_merged.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Calories', y='TotalIntensity', data=hourly_merged)
plt.title('Scatter plot of Calories vs. Total Intensity')
plt.xlabel('Calories')
plt.ylabel('Total Intensity')
plt.grid(True)
plt.show()
```
